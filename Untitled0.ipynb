```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('netflix_titles.csv')

# Inspect
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicates:", df.duplicated().sum())
print("\nFirst 5 rows:\n", df.head())
```

```python
# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['date_added'].fillna('Unknown', inplace=True)

# Fix date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Remove irrelevant columns
df.drop(columns=['show_id'], inplace=True)

# Confirm cleaning
print("✅ Cleaning Done!")
print("\nMissing values after cleaning:\n", df.isnull().sum())
print("\nNew Shape:", df.shape)
```

```python
print("Q1: Content Type Distribution")
print(df['type'].value_counts())
```

```python
print("Q2: Top 10 Countries Producing Content")
print(df['country'].value_counts().head(10))
```

```python
print("Q3: Most Common Ratings on Netflix")
print(df['rating'].value_counts().head(10))
```

```python
print("Q4: Content Added by Year")
print(df['year_added'].value_counts().sort_index())
```

```python
print("Q5: Top 10 Most Popular Genres")
from collections import Counter
genres = df['listed_in'].str.split(', ')
all_genres = [g for sublist in genres for g in sublist]
genre_counts = Counter(all_genres)
print(pd.Series(genre_counts).sort_values(ascending=False).head(10))
```

```python
plt.figure(figsize=(10, 5))
df['type'].value_counts().plot(kind='bar', color=['#E50914', '#221F1F'])
plt.title('Movies vs TV Shows on Netflix', fontsize=16)
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df['type'].value_counts().plot(kind='pie', autopct='%1.1f%%',
    colors=['#E50914', '#221F1F'], startangle=90)
plt.title('Content Type Share', fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['year_added'].value_counts().sort_index().plot(kind='line',
    marker='o', color='#E50914')
plt.title('Content Added to Netflix Per Year', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['release_year'].plot(kind='hist', bins=30, color='#E50914', edgecolor='black')
plt.title('Distribution of Release Years', fontsize=16)
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df['country'].value_counts().head(10).plot(kind='barh', color='#E50914')
plt.title('Top 10 Countries Producing Netflix Content', fontsize=16)
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pivot = df.groupby(['year_added', 'month_added']).size().unstack(fill_value=0)
sns.heatmap(pivot, cmap='Reds', linewidths=0.5)
plt.title('Heatmap of Content Added by Month & Year', fontsize=16)
plt.xlabel('Month')
plt.ylabel('Year')
plt.tight_layout()
plt.show()
```

## 📊 Insights Report

**Insight 1 — Movies Dominate Netflix**
About 70% of Netflix content is Movies and only 30% is TV Shows.
This shows Netflix focuses more on films than series.
(Reference: Chart 1 & Chart 2)

**Insight 2 — USA Leads Content Production**
United States produces the most Netflix content by far,
followed by India and United Kingdom.
This shows Netflix's strong focus on English language content.
(Reference: Chart 5)

**Insight 3 — Netflix Grew Rapidly 2016 to 2019**
Content additions peaked around 2019 showing Netflix's
aggressive expansion phase. After 2020 growth slowed
possibly due to COVID-19.
(Reference: Chart 3)

**Insight 4 — Most Content is Recent**
Majority of titles were released after 2000.
This means Netflix focuses on modern and recent content
rather than classic old movies.
(Reference: Chart 4)

**Insight 5 — December is Peak Month for Adding Content**
The heatmap shows Netflix adds the most content in
December and January, likely to attract viewers
during holiday season.
(Reference: Chart 6)

## 😲 Most Surprising Finding
I was surprised that India ranks 2nd in content production
on Netflix. This shows how strongly Netflix has invested
